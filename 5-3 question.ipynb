## 練習

```python
import pandas as pd
chipo = pd.read_csv('https://raw.githubusercontent.com/Code-Gym/python-dataset/master/chipotle.tsv',
                   sep='\t')
chipo
```

## 欄位的資料型態

```python
chipo.dtypes
```

## 將 item_price 欄位型態從「字串」轉為浮點數Float型態

```python
for value in chipo.item_price:
    print(value[1:])
```

```python
#[運算式 for 迭代變數 in 迭代器]
#float(value[1:]) 是要執行的運算式，而 chipo.item_price 是迭代器。

price = [float(value[1:]) for value in chipo.item_price]
chipo.item_price = price
    
    # price = []
    # for value in chipo.item_price:
    #     price.append(float(value[1:]))
```

## 計算 item_price 平均 (四捨五入到整數)

```python
round(chipo.item_price.mean())
```

## 請找出 item_price 價格大於 7 的資料

```python
chipo[chipo.item_price>7]
```

## 請找出 item_price 價格大於 7 ，並且 choice_description 欄位包含'Fresh Tomato Salsa' 資料
Series.str.contains('比對字串')<br>
chipo[ (條件式1) & (條件式2) ]

```python
chipo[(chipo.item_price>7)&(chipo.choice_description.str.contains('Fresh Tomato Salsa'))]
```

## 請找出 item_price最貴的一筆資料

```python
chipo.sort_values('item_price',ascending=False).head(1)
```
