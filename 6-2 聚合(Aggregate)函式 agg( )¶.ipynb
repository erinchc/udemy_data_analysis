# 就業資料分析
## 聚合(Aggregate)函式 agg( )

資料下載網址：https://raw.githubusercontent.com/Code-Gym/python-dataset/master/u.user.txt

```python
import pandas as pd
users = pd.read_csv('https://raw.githubusercontent.com/Code-Gym/python-dataset/master/u.user.txt',
                   sep='|',
                   index_col='user_id')
users
```

## 取得每一個職業的年齡平均值

```python
users.groupby('occupation').age.mean()
```

## 使用函式 agg( ) 取得每一個職業的年齡平均值
欄位.agg('函式')

```python
users.groupby('occupation').age.agg('mean')
```

## 使用函式 agg( )  取得每一個職業的年齡最小值、最大值、平均值和數量

```python
users.groupby('occupation').age.agg(['min','max','mean','count'])
```

## 將欄位名稱和函式名稱放入agg()參數，取得每一個職業的年齡平均值 
agg({'欄位':'函式'})

```python
users.groupby('occupation').agg({'age':'mean'})
```

## 將欄位名稱和函式名稱放入agg()參數，取得每一個職業和性別的年齡平均值與數量
agg({'欄位1':'函式1', '欄位2':'函式2'})

```python
users.groupby(['occupation','gender']).agg({'age':'mean', 'gender':'count'})
```
