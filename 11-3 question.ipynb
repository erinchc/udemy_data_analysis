# 練習
葡萄酒資料網址：https://raw.githubusercontent.com/Code-Gym/python-dataset/master/wine.csv

```python
import pandas as pd
import numpy as np

wine = pd.read_csv('https://raw.githubusercontent.com/Code-Gym/python-dataset/master/wine.csv')
wine
```

### 移除掉0,3,6,8,11,12,13欄位
wine.columns\[ \[0,3,6,8,11,12,13\] \]

```python
wine.drop(wine.columns[[0,3,6,8,11,12,13]],axis=1,inplace=True)
wine
```

### 建立欄位名稱
\['alcohol', 'hue', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'malic_acid'\]

```python
wine.columns=['alcohol', 'hue', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'malic_acid']
wine
```

### 設定空值
1. 請將第一欄alcohol，索引值1和2的資料改為空值
2. 請將第二欄hue，索引值3和4的資料改為空值

```python
wine.alcohol.iloc[1:3]=np.nan
wine.hue.iloc[3:5]=np.nan
wine
```

### 將欄位alcohol的空值轉為0

```python
wine.alcohol.fillna(0,inplace=True)
wine.head()
```

### 將DataFrame中剩餘的空值刪除
刪除有空值的index資料

```python
wine.dropna(axis=0,how='any',inplace=True)
wine.head()
```

### 請重置index

```python
wine=wine.reset_index(drop=True)
wine
```
