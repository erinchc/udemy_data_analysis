# 實作股價K線圖
台積電 TSMC (2330) <br>
聯發科 MediaTek (2454) 

```python
import pandas as pd
import datetime as dt
import yfinance as yf

start = '2020-01-01'
end = '2020-06-01'
symbol_1 = '2330.TW'
symbol_2 = '2454.TW'

df_2330 = yf.download(symbol_1, start=start, end=end)
df_2454 = yf.download(symbol_2, start=start, end=end)
print(df_2330.head())
print(df_2454.head())
```

### 台積電 VS 聯發科 股價線圖

```python
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(10,6))
plt.plot(df_2330.Close, label='TSMC')
plt.plot(df_2454.Close, label='MediaTek')
plt.title('TSMC vs MediaTek', loc='center')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True,axis='y') #圖表格線
plt.legend()  #圖例
plt.show()    #繪製圖表### 台積電 VS 聯發科 股價線圖
```

### 台積電股價K線圖
繪製股價圖表套件 mplfinance <br>
Github 網址：https://github.com/matplotlib/mplfinance

```python
#mav：移動平均線，將特定時間內的股價計算平均值
#volume：交易量
import mplfinance as mpf

mpf.plot(df_2330, type='candle',mav=10, volume=True)
```

```python
mpf.plot(df_2454,type='candle',mav=(3,6,9),volume=True)
```
